```python
from flask import Flask, render_template, request, jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
#searchString = request.form['content'].replace(" ","")
searchString = "Oppo9A"
flipkart_url = "https://www.flipkart.com/search?q=" + searchString
uClient = uReq(flipkart_url)
flipkart_page = uClient.read()
uClient.close()
```

```python
flipkart_html = bs(flipkart_page,"html.parser")
```

```python
bigboxes = flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"})
```

```python
del bigboxes[0:3]
```

```python
box = bigboxes[0]
box
```

```python
productlink = "https://www.flipkart.com" + box.div.div.div.a['href']
productlink
```

```python
productlink
```

```python
prodRes = requests.get(productlink)
prodRes.encoding='utf-8'
prod_html = bs(prodRes.text, "html.parser")
print(prod_html)
```

```python
commentboxes_ = prod_html.find_all('div', {'class': "_16PBlm"})
len(commentboxes_)
```

```python
commentboxes_1 = prod_html.findAll('div', {'class': "_16PBlm"})
len(commentboxes_1)
```

```python
for commentbox in commentboxes_1:
    try:
        #name.encode(encoding='utf-8')
        name = commentbox.div.div.find_all('p', {'class': '_2sc7ZR _2V5EHH'})[0].text

    except:
        name = 'No Name'

    
    try:
        comment = commentbox.div.find_all('div', {'class': ''})[0].text

    except:
        comment = 'No Comment'
        
    try:
        header = commentbox.div.find_all('p', {'class': '_2-N8zT'})[0].text

    except:
        header = 'No Comment'
    
    try:
        rating = commentbox.div.find_all('div', {'class': '_3LWZlK _1BLPMq'})[0].text

    except:
        rating = 'No Comment'
    
    
    print("Name : " + name)
    print("Product Header : " + header)
    print("Product Comment : " + comment)
    print("Rating : " + rating)
```
